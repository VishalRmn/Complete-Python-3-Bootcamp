```python
import re
```

```python
pattern = r'\d{3}-\d{3}-\d{4}' #regular exp to find phone number
```

```python
for folder, sub_folders, files in os.walk("advanced exercise puzzle"):
    print("Currently looking at folder: "+ folder)
    print('\n')
    print("THE SUBFOLDERS ARE: ")
    for sub_fold in sub_folders:
        print("\t Subfolder: "+sub_fold )
    
    print('\n')
    
    print("THE FILES ARE: ")
    for f in files:
        print("\t File: "+f)
        print('\nSearching for phone number:')
        file = open(os.path.join(folder, f), 'r') #here folder acts as the address of the path for the file, join will append the filename at the end of the path
        text = file.read()
        #print(text)
        file.close()
        results = re.search(r'\d{3}-\d{3}-\d{4}',text)
        print(results)
    print('\n')
```
